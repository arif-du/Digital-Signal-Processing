# **Even and Odd part of a Signal**

Even signal definition: When a signal exhibits symmetry with respect to t=0, then it is called an
even signal. The even signal will satisfy the condition, x(−n) = x(n).

image.png

Odd signal definition: When a signal exhibits antisymmetry with respect to t=0, then it is called an
odd signal. The odd signal will satisfy the condition, x(−n) = −x(n)

image.png

Folded Signal definition: The signal x(n) that flips around n = 0 is called folded signal which is
denoted by x(−n).

```python
n=np.arange(-6,6,1) # Specify the range of n
A=0.8
x1=A**n #Generate the given signal
x2=A**(-n) # Generate the folded signal
#Check whether the signal is odd or even
if (x2==x1).all():
  print('Even signal')
elif (x2==(-x1)).all():
  print('Odd signal')
else:
  print('Neither even nor Odd signal')
```

```python
xe=(x1+x2)/2 #Compute even part
xo=(x1-x2)/2 #Compute odd part
```

```python
#Represent the even and odd part graphically
plt.figure(figsize=(10, 8))
plt.suptitle("Even and Odd part of a signal", fontsize=10)

plt.subplot(1, 2, 1)
plt.stem(n, xe,'m', label='Even Part')
plt.xlabel('discrete time', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)
plt.legend(fontsize=10, loc='upper right')

plt.subplot(1, 2, 2)
plt.stem(n, xo, 'm', label='Odd Part')
plt.xlabel('discrete time', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)
plt.legend(fontsize=10, loc='upper right')
```
