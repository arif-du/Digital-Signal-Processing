SciPy in Python is an open-source library used for solving mathematical, scientific, engineering, and technical problems. It allows users to manipulate the data and visualize the data using a wide range of high-level Python commands.

linspace is an in-built function in Python's NumPy library. It is used to create an evenly spaced sequence in a specified interval.

The arange() function is used to get evenly spaced values within a given interval. Values are generated within the half-open interval [start, stop]. For integer arguments the function is equivalent to the Python built-in range function, but returns an ndarray rather than a list.

```python
import numpy as np
import matplotlib.pyplot as plt

import scipy
from scipy import signal

f = 20 # Hz
t = np.linspace(0, 0.5, 200)
x1 = np.sin(2 * np.pi * f * t)

s_rate = 35 # Hz. Here the sampling frequency is less than the requirement of sampling theorem

T = 1 / s_rate
n = np.arange(0, 0.5 / T)
nT = n * T
x2 = np.sin(2 * np.pi * f * nT) # Since for sampling t = nT.
```

```python
print(len(t))
print(len(nT))
```

suptitle() add a centered suptitle to the figure.

The line width simply helps to give a unique description or visual to a particular data set in a plot. The default value for the line width of a plot in Matplotlib is 1 . However, you can change this size to any size of your choice.

A shortcut for simple formatting is to use the third argument string. 'b-' means blue line, 'ro' means red circles, 'g+-' means green + with a line.

```python
plt.figure(figsize=(10, 8))
plt.suptitle("Sampling a Sine Wave of Fmax=20Hz with fs=35Hz", fontsize=10)

plt.subplot(2, 2, 1)
plt.plot(t, x1, linewidth=2, label='SineWave of frequency 20 Hz')
plt.xlabel('time.', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)
plt.legend(fontsize=10, loc='upper right')

plt.subplot(2, 2, 2)
plt.plot(nT, x2, 'ro', label='Sample marks after resampling at fs=35Hz')
plt.xlabel('time.', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)
plt.legend(fontsize=10, loc='upper right')

plt.subplot(2, 2, 3)
plt.stem(nT, x2, 'm', label='Sample after resampling at fs=35Hz')
plt.xlabel('time.', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)
plt.legend(fontsize=10, loc='upper right')

plt.subplot(2, 2, 4)
plt.plot(nT, x2, 'g-', label='Reconstructed Sine Wave')
plt.xlabel('time.', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)
plt.legend(fontsize=10, loc='upper right')

plt.tight_layout()
```
