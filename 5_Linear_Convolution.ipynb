# **Linear Convolution**

Linear Convolution is a means by which one may relate the output and input of an LTI system given the system's impulse response. Clearly, it is required to convolve the input signal with the impulse response of the system.

In linear systems, convolution is used to describe the relationship between three signals of interest: the input signal, the impulse response, and the output signal.

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
#defining sequences
x = np.array([1,2,1,-1])
h = np.array([1,2,3,1])
```

```python
#Take input sequence from user
x = input('Enter the input sequence x[n]=')
h = input('Enter the input sequence h[n]=')
```

```python
#perform convolution
y=signal.convolve(h,x)
```

```python
#defining discrete time index of two signals along with convoluted signal
#length of y(n)=length of x(n)+length of h(n)-1
n1=np.array([-1,0,1,2]) #np.arange(-2,3,1)
n2=np.array([0,1,2,3]) #np.arange(0,3,1)
n3=np.array([-1,0,1,2,3,4,5])
```

```python
#visual representation of the signals
plt.figure(figsize=(10, 8))
plt.suptitle("Convolution of two discrete sequence", fontsize=10)

plt.subplot(3, 1, 1)
plt.stem(n1, x,'m', label='input sequence')
plt.xlabel('discrete time', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)
plt.legend(fontsize=10, loc='upper right')

plt.subplot(3, 1, 2)
plt.stem(n2, h, 'm', label='inpulse sequence')
plt.xlabel('discrete time', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)
plt.legend(fontsize=10, loc='upper right')

plt.subplot(3, 1, 3)
plt.stem(n3, y, 'm', label='After convolution')
plt.xlabel('discrete time', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)
plt.legend(fontsize=10, loc='upper right')

plt.tight_layout()
```
